```python
'''
Exercise 1

    Replacing a loop
'''

def categorize_loop(temps):
    result = []
    for t in temps:
        if t < 15:
            result.append(0)  # cold
        else:
            result.append(1)  # warm
    return result
```

```python
'''
Exercise 2

    Capping Values
'''

def clamp_values(data, min_val, max_val):
    result = []
    for x in data:
        if x < min_val:
            result.append(min_val)
        elif x > max_val:
            result.append(max_val)
        else:
            result.append(x)
    return result
```

```python
'''
Exercise 3

    Accumulation
'''

def count_streak_breaks(values, threshold):
    breaks = []
    for val in values:
        if val < threshold:
            breaks.append(1)  # streak broken
        else:
            breaks.append(0)  # streak continues
    return sum(breaks)
```
